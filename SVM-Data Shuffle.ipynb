```python
# Specify the destination directories for the testing directories
glioma_test_dir = 'E:\\AI Masterclass\\MRI\\Testing\\glioma'
meningioma_test_dir = 'E:\\AI Masterclass\\MRI\\Testing\\meningioma'
pituitary_test_dir = 'E:\\AI Masterclass\\MRI\\Testing\\pituitary'
no_tumor_test_dir = 'E:\\AI Masterclass\\MRI\\Testing\\no_tumor'
```

```python
# Define the sizes for each set (train, validation, test)
train_size = 0.8  # 80% of the dataset for training
test_size = 0.2   # 20% of the dataset for testing
```

```python
# Get the list of all files in the source directory
glioma_file_list = os.listdir(glioma_source_dir)
meningioma_file_list = os.listdir(meningioma_source_dir)
pituitary_tumor_file_list = os.listdir(pituitary_source_dir)
no_tumor_file_list = os.listdir(no_tumor_source_dir)
```

```python
# Shuffle the file list randomly
random.shuffle(glioma_file_list)
random.shuffle(meningioma_file_list)
random.shuffle(pituitary_tumor_file_list)
random.shuffle(no_tumor_file_list)
```

```python
# Calculate the number of files for glioma tumors
glioma_num_files = len(glioma_file_list)
glioma_num_train = int(train_size * glioma_num_files)
glioma_num_test = int(test_size * glioma_num_files)
```

```python
# Calculate the number of files for glioma tumors
meningioma_num_files = len(meningioma_file_list)
meningioma_num_train = int(train_size * meningioma_num_files)
meningioma_num_test = int(test_size * meningioma_num_files)
```

```python
# Calculate the number of files for pituitary_tumor tumors
pituitary_tumor_num_files = len(pituitary_tumor_file_list)
pituitary_tumor_num_train = int(train_size * pituitary_tumor_num_files)
pituitary_tumor_num_test = int(test_size * pituitary_tumor_num_files)
```

```python
# Calculate the number of files for glioma tumors
no_tumor_num_files = len(no_tumor_file_list)
no_tumor_num_train = int(train_size * no_tumor_num_files)
no_tumor_num_test = int(test_size * no_tumor_num_files)
```

```python
# Split the glioma file list into train and test sets
glioma_train_files = glioma_file_list[:glioma_num_train]
glioma_test_files = glioma_file_list[glioma_num_train:]
```

```python
# Split the meningioma file list into train and test sets
meningioma_train_files = meningioma_file_list[:meningioma_num_train]
meningioma_test_files = meningioma_file_list[meningioma_num_train:]
```

```python
# Split the pituitary_tumor file list into train and test sets
pituitary_tumor_train_files = pituitary_tumor_file_list[:pituitary_tumor_num_train]
pituitary_tumor_test_files = pituitary_tumor_file_list[pituitary_tumor_num_train:]
```

```python
# Split the no_tumor file list into train and test sets
no_tumor_train_files = no_tumor_file_list[:no_tumor_num_train]
no_tumor_test_files = no_tumor_file_list[no_tumor_num_train:]
```

```python
# Move glioma files to the corresponding destination directories
for glioma_file in glioma_train_files:
    glioma_src_path = os.path.join(glioma_source_dir, glioma_file)
    glioma_dst_path = os.path.join(glioma_train_dir, glioma_file)
    shutil.copy(glioma_src_path, glioma_dst_path)


for glioma_file in glioma_test_files:
    glioma_src_path = os.path.join(glioma_source_dir, glioma_file)
    glioma_dst_path = os.path.join(glioma_test_dir, glioma_file)
    shutil.copy(glioma_src_path, glioma_dst_path)
```

```python
# Move meningioma files to the corresponding destination directories
for meningioma_file in meningioma_train_files:
    meningioma_src_path = os.path.join(meningioma_source_dir, meningioma_file)
    meningioma_dst_path = os.path.join(meningioma_train_dir, meningioma_file)
    shutil.copy(meningioma_src_path, meningioma_dst_path)


for meningioma_file in meningioma_test_files:
    meningioma_src_path = os.path.join(meningioma_source_dir, meningioma_file)
    meningioma_dst_path = os.path.join(meningioma_test_dir, meningioma_file)
    shutil.copy(meningioma_src_path, meningioma_dst_path)
```

```python
# Move pituitary_tumor files to the corresponding destination directories
for pituitary_file in pituitary_tumor_train_files:
    pituitary_src_path = os.path.join(pituitary_source_dir, pituitary_file)
    pituitary_dst_path = os.path.join(pituitary_train_dir, pituitary_file)
    shutil.copy(pituitary_src_path, pituitary_dst_path)


for pituitary_file in pituitary_tumor_test_files:
    pituitary_src_path = os.path.join(pituitary_source_dir, pituitary_file)
    pituitary_dst_path = os.path.join(pituitary_test_dir, pituitary_file)
    shutil.copy(pituitary_src_path, pituitary_dst_path)
```

```python
# Move no_tumor files to the corresponding destination directories
for no_tumor_file in no_tumor_train_files:
    no_tumor_src_path = os.path.join(no_tumor_source_dir, no_tumor_file)
    no_tumor_dst_path = os.path.join(no_tumor_train_dir, no_tumor_file)
    shutil.copy(no_tumor_src_path, no_tumor_dst_path)


for no_tumor_file in no_tumor_test_files:
    no_tumor_src_path = os.path.join(no_tumor_source_dir, no_tumor_file)
    no_tumor_dst_path = os.path.join(no_tumor_test_dir, no_tumor_file)
    shutil.copy(no_tumor_src_path, no_tumor_dst_path)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import os

path = os.listdir('E:\\AI Masterclass\\MRI\\Training')
classes = {'no_tumor':0, 'pituitary':1, 'glioma':2, 'meningioma':3} # python dic with classification clases

#The os module in Python provides a way to interact with the operating system 
```

```python
import cv2 #The cv2 module in Python is part of the OpenCV (Open Source Computer Vision) library. 
X = []   # Empty python list
Y = []
for cls in classes:
    pth = 'E:\\AI Masterclass\\MRI\\Training\\'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j,0)  # 0 is grayscale mode, Opencv default apply BGR
        img = cv2.resize(img, (200,200))
        X.append(img)
        Y.append(classes[cls])
```
